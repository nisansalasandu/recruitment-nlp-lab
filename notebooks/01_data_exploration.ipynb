## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 100)

print(" Libraries imported successfully!")
```

## 2. Load Datasets
Loading the raw resume and job description CSV files from the `data/raw/` folder.

```python
# Load datasets
resumes = pd.read_csv("../data/raw/Resume.csv")
jobs = pd.read_csv("../data/raw/job_descriptions.csv")

print(f" Resumes loaded     -> {resumes.shape[0]} rows, {resumes.shape[1]} columns")
print(f" Job Descriptions loaded -> {jobs.shape[0]} rows, {jobs.shape[1]} columns")
```

## 3. Inspect Resume Dataset

```python
print("=== RESUME DATASET ===")
print(f"Shape: {resumes.shape}")
print(f"\nColumn Names: {resumes.columns.tolist()}")
print(f"\nData Types:\n{resumes.dtypes}")
```

### 3.1 Preview First Few Rows

```python
resumes.head(5)
```

### 3.2 Check for Missing Values

```python
print("Missing values in Resumes:")
print(resumes.isnull().sum())
print(f"\nTotal missing: {resumes.isnull().sum().sum()}")
```

### 3.3 Resume Category Distribution
How many resumes exist per job category?

```python
print("Category Distribution:")
print(resumes['Category'].value_counts())
```

### 3.4 Visualize Resume Categories

```python
plt.figure(figsize=(14, 6))
resumes['Category'].value_counts().plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Resume Count by Category', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Create the folder before saving
import os
os.makedirs("../results/plots", exist_ok=True)

plt.savefig("../results/plots/resume_category_distribution.png")
plt.show()
print(" Plot saved!")
```

## 4. Inspect Job Descriptions Dataset

```python
print("=== JOB DESCRIPTIONS DATASET ===")
print(f"Shape: {jobs.shape}")
print(f"\nColumn Names: {jobs.columns.tolist()}")
print(f"\nData Types:\n{jobs.dtypes}")
```

### 4.1 Preview First Few Rows

```python
jobs.head(5)
```

### 4.2 Check for Missing Values

```python
print("Missing values in Job Descriptions:")
print(jobs.isnull().sum())
print(f"\nTotal missing: {jobs.isnull().sum().sum()}")
```

### 4.3 Job Title Distribution

```python
# Check top 20 most common job titles
print("Top 20 Job Titles:")
print(jobs['Job Title'].value_counts().head(20))
```

### 4.4 Visualize Top 20 Job Titles

```python
plt.figure(figsize=(14, 6))
jobs['Job Title'].value_counts().head(20).plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Top 20 Job Titles in Dataset', fontsize=16)
plt.xlabel('Job Title', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("../results/plots/job_title_distribution.png")
plt.show()
print(" Plot saved!")
```

## 5. Text Length Analysis
Understanding how long resumes and job descriptions are on average.

```python
# Word count analysis
resumes['word_count'] = resumes['Resume_str'].apply(lambda x: len(str(x).split()))
jobs['word_count'] = jobs['Job Description'].apply(lambda x: len(str(x).split()))

print("=== Resume Word Count ===")
print(resumes['word_count'].describe())

print("\n=== Job Description Word Count ===")
print(jobs['word_count'].describe())
```

### 5.1 Visualize Text Length Distributions

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(resumes['word_count'], bins=40, color='steelblue', edgecolor='black')
axes[0].set_title('Resume Word Count Distribution', fontsize=13)
axes[0].set_xlabel('Word Count')
axes[0].set_ylabel('Frequency')

axes[1].hist(jobs['word_count'], bins=40, color='salmon', edgecolor='black')
axes[1].set_title('Job Description Word Count Distribution', fontsize=13)
axes[1].set_xlabel('Word Count')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.savefig("../results/plots/word_count_distributions.png")
plt.show()
print(" Plot saved!")
```
